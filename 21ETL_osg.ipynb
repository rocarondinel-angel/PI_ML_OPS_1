```python
import pandas as pd
import numpy as np
import json
```

# PROCESO DE EXTRACCION

```python
# cargar datos desde un archivo JSON en un DataFrame de Pandas
df_osg = pd.read_json("output_steam_games.json", lines=True)
```

# PROCESO DE TRANSFORMACION

```python


df_osg
df_osg.info()
```

```python
df_osg.isnull().sum() 
```

```python
# eliminar registros, los cuales tengan todos sus valores nulos
df_osg = df_osg.dropna(how='all')
df_osg
# df_osg.info()
```

```python
# contar nulos
df_osg.isnull().sum() 
```

```python
# ver id nulos
df_osg[df_osg['id'].isnull()] 
```

```python
# eliminar los registros con id nulos
df_osg.dropna(subset=['id'],inplace=True) 
```

```python
# convertir a entero la columna id
df_osg['id'] = df_osg['id'].astype('int')
df_osg.info()
```

```python
# contar nulos
df_osg.isnull().sum() 
```

```python
# ver coincidencias
df_osg.sample(5)
```

```python
# imputar algunos nulos de la columna developer
df_osg['developer'].fillna(df_osg['publisher'],inplace=True)
df_osg.isnull().sum()
```

```python
# Se observan las cantidades de registros por cada fecha
df_osg['release_date'].value_counts()
```

```python
# Convertir columna release_date al tipo datetime
df_osg['release_date'] = pd.to_datetime(df_osg['release_date'], errors='coerce')
# crear columna año a partir de la columnna release_date
df_osg['año_lanzamiento'] = df_osg['release_date'].dt.year   
# modificar el tipo de la columna año_lanzamiento a entero
df_osg['año_lanzamiento'] = df_osg['año_lanzamiento'].astype('Int64')
# eliminar columna release_date
df_osg = df_osg.drop('release_date', axis=1)

df_osg
```

```python
df_osg.isnull().sum()
```

```python

df_osg['price'] = pd.to_numeric(df_osg['price'], errors='coerce')
# reemplazar cadenas con ceros
df_osg['price'] = df_osg['price'].replace(['Free to Use', 'Free to Play'], 0)
# imputar valores nulos como 0
df_osg['price'] = df_osg['price'].fillna(0)
# convertir a tipo numerico
df_osg['price'] = pd.to_numeric(df_osg['price'])
```

```python
# convertir cada valor único en la columna genres en una fila independiente.
df_osg = df_osg.explode('genres')
# eliminar registros donde el valor en la columna genres sea nulo
df_osg = df_osg.dropna(subset=['genres'])
df_osg.head(20)
```

```python
df_osg.info()
```

```python
# eliminar columnas innecesarias
df_osg = df_osg.drop(['tags', 'specs', 'url', 'reviews_url'], axis=1)
df_osg.info()
```

# PROCESO DE CARGA

```python
# convertir a csv el DataFrame df_osg
df_osg.to_csv('231_osg.csv', index=False, encoding='utf-8')
```
