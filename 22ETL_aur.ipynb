```python
import pandas as pd
import numpy as np
import json
import ast
```

# PROCESO DE EXTRACCION 

```python
# guardar cada fila en una lista
lista_aur=[]
with open("australian_user_reviews.json",encoding="utf-8") as file:
    for line in file.readlines():
        lista_aur.append(ast.literal_eval(line))
```

# PROCESO DE TRANSFORMACION 

```python
# crear dataframe
df_reviews=pd.DataFrame(lista_aur)
```

```python
# identificar la naturaleza de las columnas
df_reviews
```

```python
# ver la estructura de los datos
df_reviews.info()
```

```python
# ver si existen duplicados basandonos en la columna user_id
duplicados= df_reviews[df_reviews.duplicated(subset=['user_id'])]
duplicados
```

```python
# borrar duplicados usando la columna "user_id"
df_reviews_sin_duplicados = df_reviews.drop_duplicates(subset=['user_id'])
df_reviews_sin_duplicados
```

```python
# transformar a columnas cada elemento de las listas
df_reviews2 = pd.json_normalize(df_reviews['reviews'])
df_reviews2.head()
```

```python
# agregar columnas user_id y user_url a las columnas separadas 
df_reviews2 = pd.concat([df_reviews[['user_id', 'user_url']], df_reviews2], axis=1)
df_reviews2.head()
```

```python
# reestructurar el dataframe utilizando la funcion pd.melt teniendo como variables de identificacion a user_id y a user_url
df_reviews2 = pd.melt(df_reviews2, id_vars=['user_id', 'user_url'], 
                       value_vars=list(range(9)),
                       value_name='reviews')
```

```python
# ver las primeras filas 
df_reviews2.head()
```

```python
# eliminar las filas con valor None
df_reviews2 = df_reviews2.dropna()
```

```python
# separar por columnas cada una de las claves de la columna reviews
df_reviews = df_reviews2['reviews'].apply(pd.Series, dtype='object')
df_reviews = df_reviews.add_prefix('reviews_')
df_reviews.head()
```

```python
# concatenar con la columnas user_id y user_url
df_reviews = pd.concat([df_reviews2[['user_id', 'user_url']], df_reviews], axis=1)
df_reviews.head()
```

```python
# reemplazar los vacios con None
df_reviews.replace('', None, inplace=True)
df_reviews.head()
```

```python
# ver la nueva estructura
df_reviews.info()
```

```python
# eliminar las columnas reviews_funny y reviews_last_edited
df_reviews = df_reviews.drop(columns=['reviews_funny', 'reviews_last_edited'])
df_reviews.info()
```
